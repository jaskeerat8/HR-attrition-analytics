# Employee distribution

```python
existcount = existdata.groupby("dept")["Emp ID"].nunique().to_frame()
existcount["Ex Employees"] = exdata.groupby("dept")["Emp ID"].nunique()
existcount["dept"] = existcount.index

existcount.plot(kind = "bar",color = ["#1fed0c","#eb2005"],edgecolor = "#000000")
plt.title("Count of Employees")
plt.xlabel("Different Departments")
plt.ylabel("Count of Employees")
plt.legend(["Current Employees","Ex Employees"])
plt.grid(True)
plt.show()
```

```python
pieexist = existdata.groupby("dept")["Emp ID"].nunique().to_frame()
plt.pie(pieexist["Emp ID"],labels = pieexist.index,autopct='%1.2f%%')
plt.title("Percentage distribution of Existing Employees")
plt.xlabel("Different Departments")
plt.legend(bbox_to_anchor=(1,0), loc="center right")
plt.show()
```

```python
pieex = exdata.groupby("dept")["Emp ID"].nunique().to_frame()
plt.pie(pieex["Emp ID"],labels = pieexist.index,autopct='%1.2f%%')
plt.title("Percentage distribution of Ex Employees")
plt.xlabel("Different Departments")
plt.legend(bbox_to_anchor=(1,0), loc="center right")
plt.show()
```
